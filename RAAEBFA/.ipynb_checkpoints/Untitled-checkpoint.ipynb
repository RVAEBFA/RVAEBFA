```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# contamination ratio = 0.1
RVAE_BFA_file = '../VAESCAD/result/2019-11-20 15:35:57-BFA-kddcup.npy'
RAAE_BFA_file = './result/2019-11-20 15:35:41-BFA-kddcup.npy'
```

```python
RVAE = np.load(RVAE_BFA_file, allow_pickle=True)
RAAE = np.load(RAAE_BFA_file, allow_pickle=True)
```

```python
RVAE = RVAE.item()
RAAE = RAAE.item()
```

```python
RVAE_data = np.array(RVAE['metric'])
RAAE_data = np.array(RAAE['metric'])
x = RVAE['range']
```

```python
fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.plot(x, RVAE_data[:,0], color='#FF8C00', linestyle='-.', linewidth=2.5, marker='>', markersize = 10)
plt.plot(x, RAAE_data[:, 0], color='#1E90FF', linestyle=':', linewidth=2.5, marker='*', markersize = 10)
plt.xlabel('Balance factor $u$')
plt.ylabel('Accuracy')
plt.legend(['RVAE-BFA', 'RAAE-BFA'])
plt.savefig('./c_02.pdf')
plt.show()
```
